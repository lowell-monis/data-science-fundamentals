## Basic Data Handing


### Cleaning data

Use the following snippet to check for missing data. The rows with missing data will appear below. If there are now missing values, no rows will appear.

```python
auto[auto.isna().any(axis=1)]
```

Sometimes, missing data is not in the form of `np.nan`. In this example, it is in the form of a question mark (?). If you are able to identify what character represents missing data, you can use the following code to replace it.

```python
auto=auto.replace('?', np.nan)
```

You can now drop all rows with missing data using the following code.

```python
auto=auto.dropna()
```

You can also drop a whole column of data by using the following snippet. Note that the `axis` parameter needs to be set to 1 for columns. The default is 0 for rows.

```python
auto.drop('name', axis=1)
```

### Data type conversion

`horsepower` is of the type `object`, which makes it difficult to work with for any numerical methods. Thus, we can convert it to an integer.

```python
auto['horsepower']=auto['horsepower'].astype('int')
```

Sometimes I may need to convert numbers to strings, because a qualitative variable was wrongly encoded numerically. This can be done for the `origin` variable.

```python
convertOrigin= {1: 'American', 2:'European', 3:'Japanese'}
```

```python
auto.origin = auto.origin.apply(lambda n: convertOrigin[n])
```

### Extracting columns of specific data types.

The following code only selects columns with numeric data.

```python
auto.select_dtypes(np.number)
```

### Combining variables from different sources

I can combine data from two different variables using the `concat()` method in `pandas`. Ensure you use `axis=1` to combine on the column.

```python
x1 = auto.mpg
x2 = auto.cylinders
pd.concat([x1,x2], axis=1)
```

### Creating dummy variables for qualitative data

We will revert the string conversion for the `origin` variable here.

```python
pd.get_dummies(auto.origin, prefix='origin')
```

### Normalizing the data

Two great ways to normalize are using the standard scaling process, or by leveraging the normal distribution and z-scores directly.

#### Direct normalization using mean and standard deviation

```python
x=auto.select_dtypes(np.number)
(x-x.mean())/x.std()
```

#### Using `scikit-learn`'s standard scaling

```python
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(x).set_output(transform='pandas')
transformer.transform(x)
```

## Validation methods

### Validation set method

```python
from sklearn.model_selection import train_test_split
X=auto.drop(['mpg','origin','name'], axis=1)
y=auto[['mpg']]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=381)
```

### Leave One Out Cross-validation (LOOCV)

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

loo = LeaveOneOut()
loo.get_n_splits(X)
loo = LeaveOneOut()
model = LinearRegression()
mse_list = []
for train_index, test_index in loo.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)

np.mean(mse_list)
```

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

cv = LeaveOneOut()
model = LinearRegression()
scores = cross_val_score(model, X, y, 
                         scoring='neg_mean_squared_error',
                         cv=cv)
np.average(np.absolute(scores))
```

### K-fold Cross-validation

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
kf = KFold(n_splits=5, shuffle=True, random_state=42)
mse_list = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)
np.average(mse_list)
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression

model = LinearRegression()
kf = KFold(n_splits=5, shuffle = True)
scores = cross_val_score(model, X, y, 
                         scoring='neg_mean_squared_error',
                         cv=kf)
np.average(np.absolute(scores))
```

## KNN classification

### Model fitting

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X=auto.drop(['origin','name'], axis=1)
y=auto['origin']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
accuracy_score(knn.predict(X_train),y_train)
```

### Making predictions 

```python
accuracy_score(knn.predict(X_test),y_test)
```

```python
knn.predict_proba(X_test)
```
